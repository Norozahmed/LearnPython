### Type Conversion

Type conversion in Python refers to the process of changing a value from one data type to another. This is often necessary when you need to perform operations that are not compatible with the original data type, or when you need to represent data in a different format.

There are two main types conversion in python:
1. Implicit Type Conversion (Coercion):
* This occurs automatically when Python performs certain operations.
* For example, when you add an integer and a float, Python implicitly converts the integer to a float before performing the addition.



```python
num_int = 123
num_float = 1.23

num_new = num_int + num_float

print("datatype of num_int:",type(num_int))
print("datatype of num_float:",type(num_float))

print("Value of num_new:",num_new)
print("datatype of num_new:",type(num_new))
```

2. Explicit Type conversion (Type Casting):
* This is done manually using built-in functions.
* Python provides functions like int(), float(), str(), list(), tuple(), set(), etc., to convert values to specific data types.

```python
# int(): Converts a value to an integer.
num_str = "12"
num_int = int(num_str) # '12' is converted to integer 12
print(type(num_int)) # Output: <class 'int'>

num_float = 10.7
num_int = int(num_float) # 10.7 is converted to integer 10
print(num_int) # Output: 10
```

```python
# flaot(): Converts a value to a floating-point number.
num_str = "10"
num_float = float(num_str) # '10' is converted to float 10.0
print(type(num_float)) # Output: <class 'float'>

num_int = 12
num_float = float(num_int) # 12 is converted to float 12.0
print(num_float) # Output: 12.0
```

```python
# str(): Converts a value to a string
num_int = 10
num_str = str(num_int) # 10 is converted to string '10'
print(type(num_str)) # Output: <class 'str'>

num_float = 12.3
num_str = str(num_float) # 12.3 is converted to string '12.3'
print(num_str) # Output: '12.3'
```

***Important Consideration:***
* Not all conversions are possible. For example, you cannot directly convert a string like "hello" to an integer.
* When converting from float to int, the decimal part is truncated (not rounded).
* Type conversion can lead to loss of information. For example, converting a float to an int will discard the decimal part.

### [If/Elif/Else - Conditional Statements](https://www.w3schools.com/python/python_conditions.asp)

if, elif, and else are conditional statements in Python used to execute different blocks of code based on whether a condition is true or false. They allow your program to make decisions and follow different paths of execution.

### if statement:

The if statement is the most basic form of a conditional statement. It executes a block of code only if a specified condition is true.

if---------statement.webp

```python
'''
if condition:
    # Code to be executed if the condition is true
'''
```



* condition: An expression that evaluates to either True or False.
* The code within the indented block following the if statement is executed only if the condition is True.



```python
# Example:
x = 10
if x > 5:
    print("x is greater than 5")  # This will be printed because 10 > 5 is True
```

### else statement:

The else statement is used in conjunction with an if statement. The code within the else block is executed if the if condition is false.


if---------else---------statement.webp

```python
'''
if condition:
    # Code to be executed if the condition is true
else:
    # Code to be executed if the condition is false
'''
```

```python
# Example:
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")  # This will be printed because 3 > 5 is False
```

### elif statement (else if):

The elif statement allows you to check multiple conditions in sequence. It's short for "else if". If the initial if condition is false, the elif condition is checked. If the elif condition is true, its corresponding block of code is executed. You can have multiple elif statements.

If-elif-else-Statement--1.webp

```python
'''
if condition1:
    # Code to be executed if condition1 is true
elif condition2:
    # Code to be executed if condition1 is false and condition2 is true
elif condition3:
    # Code to be executed if condition1 and condition2 are false and condition3 is true
else:
    # Code to be executed if all conditions are false
'''
```

[Example:](https://github.com/rameenf1024/CppSnippetsSculptor)

```python
x = 7

if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5")  # This will be printed because 7 > 5 is True
elif x > 2:
    print("x is greater than 2")
else:
    print("x is not greater than 2")
```

*Important Note* :



*   **Indentation**: Python uses indentation to define code blocks. The code within an if, elif, or else block must be indented. Consistent indentation is crucial for correct execution. Typically, 4 spaces are used for indentation.
*   **Order**: The conditions are checked in order. Once a condition is found to be true, its corresponding block of code is executed, and the rest of the elif and else blocks are skipped.
Optional else: The else block is optional. You can have an if statement with or without an else.
*  **Chained if**: You can also have nested or chained if statements (if statements inside if statements).


```python
# Example of chained if:
age = 25
has_license = True

if age >= 18:
    if has_license:
        print("Can drive")
    else:
        print("Old enough to drive, but needs a license")
else:
    print("Too young to drive")
```

Modules in Python are files containing Python code (definitions, functions, classes, and variables) that you can use in other Python programs. They allow for better organization, code reuse, and abstraction by breaking programs into smaller, manageable, and reusable pieces.



***types of Python Modules:***

* These modules are included with Python and provide standard functionality.
* Examples: os, sys, math, datetime, random, etc.

* Examples usage:


```python
import math
print(math.sqrt(16))  # Output: 4.0
```

2. User-defined Modules:


*   These are modules you create yourself by saving Python code into a file (e.g., mymodule.py) and importing it into another program.
*   Example:


```python
# mymodule.py
def greet(name):
    return f"Hello, {name}!"

# main.py
import mymodule
print(mymodule.greet("Alice"))
```

```python
# or
import random

def play_game():
    number = random.randint(1, 50)
    while True:
        try:
            guess_number = int(input("Guess a number between 1 and 50: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue  # Go back to the beginning of the loop

        if guess_number < number:
            print("Too low")
        elif guess_number > number:
            print("Too high")
        else:
            print("Correct")
            print(f"The number is {number}")
            return  # Exit the game loop


while True:
    play_game()
    ask1 = input("Do you want to play again? (yes / no): ").lower()
    if ask1 != "yes":
        break  # Exit the outer loop and end the game
```

here is a breakdown:
* Function for Gameplay: The core game logic is now encapsulated within the play_game() function. This avoids code duplication when the user wants to play again.
* while True Loops: Using while True loops for both the game itself and the play-again prompt makes the control flow cleaner. We use break statements to exit these loops when needed.
* return to End Game: Inside the play_game() function, when the user guesses correctly, the return statement exits the function, simplifying the logic.
* Input Validation: The try...except block handles potential ValueError exceptions that occur if the user enters non-numeric input. This makes the game more robust.
* Removed Redundant int(): The original code had int(int(input(...))). The outer int() was unnecessary and has been removed.
* Simplified Play Again Logic: The play again logic is much more concise. The outer while True loop keeps the game going until the user enters something other than "yes".


These changes result in cleaner, more efficient, and easier-to-read code. It also makes it easier to extend the game in the future (e.g., adding a limited number of guesses).

### [Nested For Loop](https://www.geeksforgeeks.org/python-nested-loops/)

In Python programming language there are two types of loops which are for loop and while loop. Using these loops we can create nested loops in Python. Nested loops mean loops inside a loop. For example, while loop inside the for loop, for loop inside the for loop, etc.

[github.com](https://github.com/faisal2410/ostad_basic_b1)

Nestedloop.png

```python
'''
for outer_loop_variable in outer_sequence:
    for inner_loop_variable in inner_sequence:
        # Code to be executed
'''
```

```python
# example

for i in range(3):
  for j in range(2):
    print(f"i: {i}, j: {j}")
```

**Explanation**

* the outer loop iterates three times (i = 0, 1, 2).
* For each value of *i*, the inner loop iterates twice (i = 0, 1).
* The code inside the inner loop is executed for each combination of *i* and *j*.

**Use Cases:**

Nested for loops are commonly used for:
* **Iterating over 2D data structure:** Like lists of lists (matrics) or grids.
* **Generating combinations or permutations:** Of elements from multiple sequences.  
* **Solve certain algorithmic problems:** That require nested iterations.

```python
# Example: Full pyramid patterns

n = int(input("Enter a number: "))  # Number of rows

for i in range(1, n + 1):  # Outer loop for rows
    # Print spaces
    for j in range(n - i):  # Inner loop for spaces
        print(" ", end="")

    # Print stars
    for k in range(2 * i - 1):  # Inner loop for stars
        print("*", end="")

    print() # Newline after each row
```

```python
# Full pyramid patterns in python recurrsion

def print_space(space):
    if space > 0:
        print(" ", end="")
        print_space(space - 1)

def print_star(star):
    if star > 0:
        print("*", end="")
        print_star(star - 1)

def print_pyramid(n, current_row=1):
    if current_row > n:
        return

    spaces = n - current_row
    stars = 2 * current_row - 1

    # Print spaces
    print_space(spaces)

    # Print stars
    print_star(stars)

    # Move to the next line for the next row
    print()

    # Print the pyramid for the next row
    print_pyramid(n, current_row + 1)

# Set the number of rows for the pyramid
n = 5

# Print the pyramid pattern
print_pyramid(n)

```

### [Loop Control Statement](https://www.geeksforgeeks.org/loops-and-control-statements-continue-break-and-pass-in-python/)

Loop control statements are essential tools in programming that allow you to alter the flow of execution within loops. They enable you to skip iterations, terminate loops prematurely, or jump to specific points within the code.
In Python, loop control statements are used to change the execution sequence of a loop. By default, loops execute until their condition becomes false or until all items in a sequence are iterated. However, loop control statements allow you to alter this default behavior.

There are three main loop control statements in Python:

1. **break statement**:

* The *break* statement is used to terminate the loop prematurely. When a *break* statement is encountered inside a loop, the loop is immediately exited, and the program control is transferred to the statement immediately following the loop.

```python
for i in range(1, 11):
  if i == 5:
    break   # Exit the loop when i is 5
  print(i)
# Output: 1 2 3 4
```

* The *continue* statement is used to skip the rest of the current iteration and proceed to the next iteration of1 the loop. When a continue statement is encountered, the remaining statements within the loop's body for the current iteration are skipped, and the loop continues with the next iteration.

Logical operators (and, or, not) can be used with strings in the context of boolean evaluations (e.g., in if statements). However, they don't directly manipulate the strings themselves.  

* A non-empty string is considered "truthy" (evaluates to True in a boolean context).

### [f-String](https://www.w3schools.com/python/python_string_formatting.asp)    

F-strings, or formatted string literals, are a concise and powerful way to embed expressions inside string literals in Python. Introduced in Python 3.6, they provide a more readable and efficient alternative to the older %-formatting and .format() methods.    

**How F-strings Work**   

You create an f-string by prefixing a string literal with the letter f or F. Inside the string, you can place expressions within curly braces {}, and these expressions will be evaluated and their values inserted into the string.


```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")  # Output: My name is Alice and I am 30 years old.
```

```python
x = 10
y = 5
print(f"The sum of {x} and {y} is {x + y}.")  # Output: The sum of 10 and 5 is 15.

# Calling functions
def greet(name):
    return f"Hello, {name}!"

print(greet("Bob"))  # Output: Hello, Bob!

# Accessing dictionary values
person = {"name": "Charlie", "city": "London"}
print(f"{person['name']} lives in {person['city']}.")  # Output: Charlie lives in London.
```

*Formatting Specifiers*    

F-strings also support format specifiers, similar to the .format() method, allowing you to control the appearance of the inserted values.

```python
pi = 3.14159265359
print(f"Pi rounded to two decimal places: {pi:.2f}")  # Output: Pi rounded to two decimal places: 3.14

price = 19.99
print(f"The price is ${price:.2f}")  # Output: The price is $19.99

import datetime
now = datetime.datetime.now()
print(f"Today's date is: {now:%Y-%m-%d}")  # Output (example): Today's date is: 2024-07-24
```

***Benefits of F-strings***  
* **Readability:** F-strings make string formatting more intuitive and easier to read, especially when dealing with complex expressions.
* **Conciseness:** They often require less code than the older formatting methods.      
* **Efficiency:** F-strings are generally faster than %-formatting and .format().  
* **Direct Expression Evaluation:** You can directly embed expressions without needing to create intermediate variables.

```python
# Example: Building a Table

products = [
    {"name": "Laptop", "price": 1200, "quantity": 2},
    {"name": "Mouse", "price": 25, "quantity": 5},
    {"name": "Keyboard", "price": 75, "quantity": 3},
]

print("Product  | Price | Quantity | Total")
print("-------  | ----- | -------- | -----")

for product in products:
    total = product["price"] * product["quantity"]
    print(f"{product['name']:<8} | {product['price']:>5} | {product['quantity']:^8} | {total:>5}")

# Output (formatted table):
# Product | Price | Quantity | Total
# ------- | ----- | -------- | -----
# Laptop  |  1200 |    2     |  2400
# Mouse   |    25 |    5     |   125
# Keyboard|    75 |    3     |   225
```
